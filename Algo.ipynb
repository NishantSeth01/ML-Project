```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("laptop_data.csv")
df.head()
```

```python
df = df.drop("Unnamed: 0", axis = 1)
df.head()
```

```python
df.Company.value_counts()
```

```python
df = df.join(pd.get_dummies(df.Company))
df = df.drop("Company", axis = 1)
df
```

```python
df.TypeName.value_counts()
```

```python
df = df.join(pd.get_dummies(df.TypeName))
df = df.drop("TypeName", axis = 1)
df.head()
```

```python
df.ScreenResolution.value_counts()
```

```python
df["ScreenResolution"] = df.ScreenResolution.str.split(" ").apply(lambda x: x[-1])
df["ScreenWidth"] = df.ScreenResolution.str.split("x").apply(lambda x: x[0])
df["ScreenHeight"] = df.ScreenResolution.str.split("x").apply(lambda x: x[1])
df.head()
```

```python
df = df.drop("ScreenResolution", axis = 1)
df.head()
```

```python
df["CpuBrand"] = df.Cpu.str.split(" ").apply(lambda x: x[0])
df["CpuFreq"] = df.Cpu.str.split(" ").apply(lambda x: x[-1])
df = df.drop("Cpu", axis = 1)
df.head()
```

```python
df.CpuBrand.value_counts()
```

```python
sn.countplot(df["CpuBrand"])
plt.show()
```

```python
df["CpuFreq"] = df.CpuFreq.str[:-3]
df.head()
```

```python
df["Ram"] = df.Ram.str[:-2].astype("int")
df.head()
```

```python
sn.countplot(y = df.Ram)
plt.show()
```

```python
df["CpuFreq"] = df["CpuFreq"].astype("float")
```

```python
df.head()
```

```python
df["ScreenWidth"] = df.ScreenWidth.astype("int")
df["ScreenHeight"] = df.ScreenHeight.astype("int")
```

```python
df[["ScreenWidth", "ScreenHeight", "Inches", "Ram", "CpuFreq"]].hist()
```

```python
df["Storage"] = df.Memory.str.split(" ").apply(lambda x: x[0])
df["MemoryType"] = df.Memory.str.split(" ").apply(lambda x: x[1])
df.head()
```

```python
df.MemoryType.value_counts()
```

```python
sn.countplot(df["MemoryType"])
plt.show()
```

```python
df["Storage"].value_counts()
```

```python
def storage_convert(x):
    if "GB" in x:
        return float(x[: x.find("GB")])
    elif "TB" in x:
        return float(x[: x.find("TB")]) * 1000
df["Storage"] = df["Storage"].apply(storage_convert)
df.head()
```

```python
df.Storage.value_counts()
```

```python
df = df.drop("Memory", axis = 1)
df.head()
```

```python
df["Weight"] = df.Weight.str[: -2].astype("float")
df.Weight.value_counts()
```

```python
df.Gpu.value_counts()
```

```python
df["GpuBrand"] = df.Gpu.str.split(" ").apply(lambda x: x[0])
df.GpuBrand.value_counts()
```

```python
sn.countplot(df["GpuBrand"])
plt.show()
```

```python
df = df.drop("Gpu", axis = 1)
df.head()
```

```python
df = df.join(pd.get_dummies(df["OpSys"]))
df.head()
```

```python
df = df.drop("OpSys", axis = 1)
df.head()
```

```python
df.info()
```

```python
sn.heatmap(df.corr())
plt.show()
```

```python
cpu_cat = pd.get_dummies(df["CpuBrand"])
cpu_cat.columns = [col + "_CPU" for col in cpu_cat.columns]
gpu_cat = pd.get_dummies(df["GpuBrand"])
gpu_cat.columns = [col + "_GPU" for col in gpu_cat.columns]

df = df.join([cpu_cat, gpu_cat])
df.drop(["CpuBrand", "GpuBrand"], axis = 1)
df.head()
```

```python
df.info()
```

```python
df.corr()
```

```python
plt.figure(figsize = (10, 10))
sn.heatmap(df.corr())
plt.show()
```

```python
tar = df.corr()["Price"].apply(abs).sort_values()
tar
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

X = df[list(tar.index)].drop("Price", axis = 1)
y = df["Price"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15)

scale = StandardScaler()
X_train_scaled = scale.fit_transform(X_train)
X_test_scaled = scale.transform(X_test)
```

```python
rf = RandomForestRegressor()

rf.fit(X_train_scaled, y_train)
```

```python
rf.score(X_test_scaled, y_test)
```

```python
y_pred = rf.predict(X_test_scaled)
```

```python
plt.figure(figsize = (12, 8))
plt.scatter(x = y_pred, y = y_test)
plt.plot(range(0, 175000), range(0, 175000), c = "r")
plt.xlabel("Predicted Value")
plt.ylabel("Actual Value")
plt.title("Predicted Value VS Actual Value")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

print("MSE: ", mean_squared_error(y_test, y_pred), "\nRMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
import pickle

with open('rf_model.pkl', 'wb') as model_file:
    pickle.dump(rf, model_file)
```
